#  Pokec Network Study (sampled version)

We‚Äôre going to:
1. Load the giant Pokec friendship graph  
2. Extract a connected 10 000-node subgraph using BFS sampling  
3. Load and clean user attributes  
4. Attach those attributes to the sampled graph  
5. Visualize it and confirm it behaves like the full network  


```python
# ==================================
#  Imports
# ==================================
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
import os

plt.style.use("seaborn-v0_8")
```

## 1Ô∏è Load a partial graph and sample a connected subgraph
The full Pokec dataset has 1.6 M nodes and 30 M edges ‚Äî too heavy.
We load the first few hundred thousand edges, then extract a BFS subgraph (~10 k nodes).


```python
import os
import networkx as nx
import random
import numpy as np

# Fixer la graine pour reproductibilit√©
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

GRAPH_PATH = "dataset/pokec_subgraph_relationships.gpickle"

def load_partial_graph(path, max_edges=500_000):
    G = nx.Graph()
    with open(path) as f:
        for i, line in enumerate(f):
            if i >= max_edges:
                break
            try:
                u, v = map(int, line.strip().split('\t'))
                G.add_edge(u, v)
            except:
                continue
    return G

# --- Sauvegarde / Chargement compatible NetworkX 3.x ---
def save_graph_gpickle(G, path):
    """Save graph using pickle (compatible with all NetworkX versions)."""
    import pickle
    with open(path, "wb") as f:
        pickle.dump(G, f)

def load_graph_gpickle(path):
    """Load graph using pickle (compatible with all NetworkX versions)."""
    import pickle
    with open(path, "rb") as f:
        return pickle.load(f)

# --- V√©rifie si le graphe est d√©j√† sauvegard√© ---
if os.path.exists(GRAPH_PATH):
    print(f"üìÇ Loading saved subgraph from {GRAPH_PATH}...")
    H = load_graph_gpickle(GRAPH_PATH)
    print(f"‚úÖ Loaded saved graph: {H.number_of_nodes()} nodes, {H.number_of_edges()} edges")

else:
    print("‚è≥ Loading part of Pokec edges...")
    G = load_partial_graph("dataset/soc-pokec-relationships.txt", max_edges=500_000)
    print(f"‚úÖ Partial graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

    # BFS sample autour d‚Äôun n≈ìud choisi al√©atoirement mais d√©terministe
    start = random.choice(list(G.nodes()))
    sub_nodes = list(nx.bfs_tree(G, start, depth_limit=3).nodes())
    H = G.subgraph(sub_nodes).copy()

    # Si le graphe est trop petit, √©largir la recherche
    if H.number_of_nodes() < 1000:
        print(f"Too small ({H.number_of_nodes()} nodes), trying deeper BFS...")
        sub_nodes = list(nx.bfs_tree(G, start, depth_limit=4).nodes())
        H = G.subgraph(sub_nodes).copy()

    print(f"‚úÖ BFS subgraph: {H.number_of_nodes()} nodes, {H.number_of_edges()} edges")

    # üíæ Sauvegarder avec pickle
    os.makedirs(os.path.dirname(GRAPH_PATH), exist_ok=True)
    save_graph_gpickle(H, GRAPH_PATH)
    print(f"üíæ Saved subgraph to {GRAPH_PATH}")
```

## 2Ô∏è Load and clean user features


##  Phase Pr√©liminaire ‚Äî Pr√©-traitement des donn√©es avant toute analyse GNN

Avant toute analyse de graphe de type **Graph Neural Network (GNN)** sur des donn√©es r√©elles, l'√©tape cruciale est le **pr√©-traitement**.

Principe fondamental : si les donn√©es d‚Äôentr√©e sont mauvaises (*GIGO : Garbage In, Garbage Out*), les r√©sultats le seront aussi.  
Les GNN exploitent les **attributs de n≈ìuds** ; ces features doivent donc √™tre propres, encod√©es et standardis√©es.

Cette section applique les principes de *Data Mining* classiques :
1. Nettoyage des donn√©es (valeurs manquantes, doublons, erreurs)
2. Encodage des variables cat√©gorielles
3. Standardisation des num√©riques
4. Nettoyage structurel du graphe (self-loops, n≈ìuds isol√©s)
5. Sauvegarde du graphe et des features pr√™tes pour l‚Äôanalyse


```python
import pandas as pd
import numpy as np
import networkx as nx
import os

PROFILES_RAW = "dataset/soc-pokec-profiles.txt"
GRAPH_PREP_PATH = "dataset/pokec_graph_preprocessed.gpickle"
FEATURES_PREP_PATH = "dataset/pokec_features_preprocessed.parquet"

print("üì¶ Pr√©paration compl√®te √† partir du fichier brut Pokec...")
```

###  √âtape 1 : Lecture et nettoyage du fichier brut

On s√©lectionne les colonnes utiles, corrige les types, et simplifie les textes libres  
(`smoker`, `drinker`) pour obtenir des cat√©gories coh√©rentes.


```python
profiles = pd.read_csv(PROFILES_RAW, sep="\t", header=None, low_memory=False)

cols = [
    "user_id","public","completion","gender","region",
    "last_login","registration","age","body","work_field",
    "languages","hobbies","interests","music","movies",
    "religion","cars","politics","looking_for","smoker","drinker"
]
profiles.columns = cols + list(range(len(cols), profiles.shape[1]))

df = profiles[["user_id","gender","age","region","smoker","drinker","completion"]].copy()
df = df.dropna(subset=["user_id"])
df["user_id"] = df["user_id"].astype(int)

# genre, r√©gion, √¢ge
df["gender"] = df["gender"].map({1.0:"male", 0.0:"female"}).fillna("unknown")
df["region"] = df["region"].astype(str).str.split(",").str[0]
df["age"] = pd.to_numeric(df["age"], errors="coerce").replace(0, np.nan)

# simplification des champs texte
def simplify_text(val):
    if not isinstance(val, str):
        return "unknown"
    val = val.lower()
    if any(k in val for k in ["ne", "nie", "vobec"]):
        return "no"
    if any(k in val for k in ["ano", "√°no", "hej", "pijem", "obcas", "obƒças"]):
        return "yes"
    return "unknown"

df["smoker_clean"] = df["smoker"].apply(simplify_text)
df["drinker_clean"] = df["drinker"].apply(simplify_text)

print("‚úÖ Profils bruts nettoy√©s :", len(df), "lignes")
print(df.head(3))
```

###  √âtape 2 : Nettoyage structurel du graphe

Avant toute extraction de features, on doit v√©rifier que le graphe n‚Äôa ni doublons ni boucles ;  
les n≈ìuds isol√©s sont supprim√©s pour ne conserver qu‚Äôune composante coh√©rente.



```python
H.remove_edges_from(nx.selfloop_edges(H))
H = nx.Graph(H)
isolates = list(nx.isolates(H))
if isolates:
    H.remove_nodes_from(isolates)
print(f"Graphe nettoy√© : {H.number_of_nodes()} n≈ìuds, {H.number_of_edges()} ar√™tes.")
```

### ‚öôÔ∏è √âtape 3 : Encodage One-Hot et standardisation (Z-score)

Les variables cat√©gorielles sont encod√©es, les num√©riques standardis√©es ;  
on obtient une matrice homog√®ne exploitable par un GNN.


```python
# Pr√©paration des features et normalisation (safe, sans chained assignment)
feat = df[df["user_id"].isin(H.nodes())].copy()

# 1) conversions num√©riques
feat["age"] = pd.to_numeric(feat["age"], errors="coerce")
feat["completion"] = pd.to_numeric(feat["completion"], errors="coerce")

# 2) imputation par la moyenne (AVANT standardisation)
feat = feat.assign(
    age=feat["age"].fillna(feat["age"].mean()),
    completion=feat["completion"].fillna(feat["completion"].mean())
)

# ...
# 3) Standardisation num√©rique
feat = feat.assign(
    age_z=(feat["age"] - feat["age"].mean()) / feat["age"].std(),
    completion_z=(feat["completion"] - feat["completion"].mean()) / feat["completion"].std()
)

# 4) Gestion des valeurs manquantes cat√©gorielles
for col in ["gender", "region", "smoker_clean", "drinker_clean"]:
    mode_value = feat[col].replace("unknown", np.nan).mode()[0]
    feat[col] = feat[col].replace("unknown", np.nan).fillna(mode_value)

# 5) Encodage One-Hot
feat_encoded = pd.get_dummies(
    feat[["gender", "region", "smoker_clean", "drinker_clean"]],
    prefix=["gender", "region", "smoke", "drink"]
)


# 5) fusion des num√©riques
feat_encoded["age_z"] = feat["age_z"]
feat_encoded["completion_z"] = feat["completion_z"]

# 6) matrice X align√©e sur l‚Äôordre des n≈ìuds
X = feat_encoded.reindex(H.nodes(), fill_value=0)

# ‚ö†Ô∏è 7) (RE)ATTACHE les attributs AUJOURD‚ÄôHUI (apr√®s imputation/standardisation et apr√®s tout subgraph)
feat_attrs = feat[[
    "user_id","gender","region","smoker","drinker","completion",
    "smoker_clean","drinker_clean","age","age_z","completion_z"
]].copy()
feat_attrs["user_id"] = feat_attrs["user_id"].astype(int)

attrs = feat_attrs.set_index("user_id").to_dict("index")
nx.set_node_attributes(H, attrs)

print(f"OK: X = {X.shape}, attrs attached to H = {H.number_of_nodes()} nodes")
```

### üíæ √âtape 4 : Extraction de la composante g√©ante et sauvegarde

On conserve la plus grande composante connexe et  
on sauvegarde le graphe et la matrice de features pour les √©tapes GNN suivantes.


```python
# ensure the dataset folder exists
os.makedirs("dataset", exist_ok=True)

# cast everything to numeric types Arrow can handle cleanly
X = X.astype("float32")

# save safely
import pickle
with open(GRAPH_PREP_PATH, "wb") as f:
    pickle.dump(H, f)

X.to_parquet(FEATURES_PREP_PATH)

print(f"üéØ Graphe sauvegard√© ‚Üí {GRAPH_PREP_PATH}")
print(f"üéØ Features sauvegard√©es ‚Üí {FEATURES_PREP_PATH}")
```

## 6Ô∏è Visualization (color by region)


```python
import random
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Pick a random starting node
start = random.choice(list(H.nodes()))

# Get neighbors up to 2 hops away
connected_nodes = list(nx.single_source_shortest_path_length(H, start, cutoff=2).keys())
S = H.subgraph(connected_nodes).copy()

print(f"Visualizing connected subgraph: {S.number_of_nodes()} nodes, {S.number_of_edges()} edges")

# Layout + visualization
pos = nx.spring_layout(S, seed=42, iterations=30)

regions = [S.nodes[n].get("region", "unknown") for n in S.nodes()]
unique_regions = sorted(set(regions))
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_regions)))
color_dict = {r: colors[i % 20] for i, r in enumerate(unique_regions)}
node_colors = [color_dict[S.nodes[n].get("region", "unknown")] for n in S.nodes()]

plt.figure(figsize=(10, 8))
nx.draw(
    S, pos,
    node_color=node_colors,
    node_size=40,
    edge_color="gray",
    with_labels=False,
    alpha=0.85
)
plt.title(f"Pokec Connected Subgraph ({S.number_of_nodes()} nodes) ‚Äî colored by region")
plt.show()
```

# Analyse Exploratoire du Graphe (AEG)

Avant toute mod√©lisation (GNN, pr√©diction de liens, etc.), on r√©alise une **analyse exploratoire du graphe**.
Cette phase permet de comprendre la structure, l‚Äôimportance des n≈ìuds et la pr√©sence √©ventuelle de communaut√©s.

Les √©tapes suivantes sont :
1. Chargement du graphe pr√©trait√©
2. Analyse des propri√©t√©s globales (macro)
3. √âtude des centralit√©s (micro)
4. D√©tection de communaut√©s et mesure d‚Äôassortativit√© (m√©so)


### Chargement du graphe

```python
import pickle
import networkx as nx
import pandas as pd

GRAPH_PREP_PATH = "dataset/pokec_graph_preprocessed.gpickle"
FEATURES_PREP_PATH = "dataset/pokec_features_preprocessed.parquet"

with open(GRAPH_PREP_PATH, "rb") as f:
    H = pickle.load(f)

X = pd.read_parquet(FEATURES_PREP_PATH)

print(f"Graphe charg√© : {H.number_of_nodes()} n≈ìuds, {H.number_of_edges()} ar√™tes")
print(f"Matrice de features : {X.shape[0]} lignes, {X.shape[1]} colonnes")
```

## 1. Propri√©t√©s Globales du Graphe

On mesure ici la **structure g√©n√©rale** du r√©seau :
- Taille (nombre de n≈ìuds et d‚Äôar√™tes)
- Densit√©
- Coefficient de clustering global
- Longueur moyenne du plus court chemin
- Diam√®tre
Ces m√©triques aident √† caract√©riser le type de r√©seau (dense, clairsem√©, petit monde, etc.).


```python
import numpy as np
import random
import networkx as nx

# 1. Propri√©t√©s de base
N = H.number_of_nodes()
M = H.number_of_edges()
density = nx.density(H)
clustering = nx.transitivity(H)

# 2. Extraction de la plus grande composante connexe
largest_cc = max(nx.connected_components(H), key=len)
H_cc = H.subgraph(largest_cc).copy()

# 3. Approximation de la longueur moyenne du plus court chemin et du diam√®tre
# On √©chantillonne un sous-ensemble de n≈ìuds pour acc√©l√©rer le calcul
sample_size = min(2000, len(H_cc))  # on limite √† 2000 n≈ìuds max
sample_nodes = random.sample(list(H_cc.nodes()), sample_size)

distances = []
for n in sample_nodes:
    # Single-source shortest path lengths for node n
    sp_lengths = nx.single_source_shortest_path_length(H_cc, n)
    distances.extend(sp_lengths.values())

# Moyenne et percentiles comme approximation
avg_path_length = np.mean(distances)
diameter_approx = np.percentile(distances, 99.9)  # estimation du "diam√®tre"

# 4. Affichage des r√©sultats
print(f"Nombre de n≈ìuds : {N}")
print(f"Nombre d‚Äôar√™tes : {M}")
print(f"Densit√© : {density:.6f}")
print(f"Clustering global : {clustering:.4f}")
print(f"Longueur moyenne approxim√©e du plus court chemin : {avg_path_length:.3f}")
print(f"Diam√®tre approxim√© (99.9e percentile) : {diameter_approx:.0f}")
```

# √âtude de la Centralit√© du Graphe

L‚Äô√©tude de la centralit√© vise √† comprendre **le r√¥le structurel des n≈ìuds** : quels sont les plus connect√©s, les plus influents ou les plus strat√©giques dans le r√©seau.

Nous allons calculer :
- **Centralit√© de degr√©** (*degree centrality*)
- **Centralit√© d‚Äôinterm√©diarit√©** (*betweenness centrality*)
- **Centralit√© de proximit√©** (*closeness centrality*)
- **PageRank**
- **Centralit√© de vecteur propre** (*eigenvector centrality*)

Toutes les mesures sont calcul√©es sur la **plus grande composante connexe** du graphe pour √©viter les biais dus aux sous-graphes isol√©s.


### Pr√©paration du graphe de travail

```python
import networkx as nx

# On travaille sur la plus grande composante connexe
largest_cc = max(nx.connected_components(H), key=len)
H_cc = H.subgraph(largest_cc).copy()

print(f"Composante utilis√©e : {H_cc.number_of_nodes()} n≈ìuds, {H_cc.number_of_edges()} ar√™tes")
```

## 1. Centralit√© de Degr√© (Degree Centrality)

La centralit√© de degr√© mesure la **popularit√©** d‚Äôun n≈ìud, c‚Äôest-√†-dire le nombre de connexions directes qu‚Äôil poss√®de.  
C‚Äôest une mesure simple, mais souvent tr√®s r√©v√©latrice dans les r√©seaux sociaux.


```python
degree_centrality = nx.degree_centrality(H_cc)
print(f"Centralit√© de degr√© calcul√©e pour {len(degree_centrality)} n≈ìuds.")

# Top 5
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
for node, score in sorted_degree:
    print(f"Node {node}: degr√© centralit√© = {score:.4f}")
```

## 2. Centralit√© d‚ÄôInterm√©diarit√© (Betweenness Centrality)

Cette mesure identifie les **n≈ìuds-ponts**, qui relient des parties du graphe entre elles.  
Elle repose sur les plus courts chemins entre tous les couples de n≈ìuds.

Comme le calcul exact est co√ªteux sur de grands graphes, on utilise ici une **approximation par √©chantillonnage**.


```python
betweenness_centrality = nx.betweenness_centrality(H_cc, k=1000, seed=42)  # approximation rapide
print(f"Centralit√© d‚Äôinterm√©diarit√© calcul√©e (approximation sur 1000 n≈ìuds).")

top_betw = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
for node, score in top_betw:
    print(f"Node {node}: betweenness = {score:.4f}")
```

## 3. Centralit√© de Proximit√© (Closeness Centrality)

Cette mesure indique la **rapidit√© avec laquelle un n≈ìud peut atteindre les autres**.  
Plus la distance moyenne vers les autres n≈ìuds est faible, plus la centralit√© est √©lev√©e.


```python
closeness_centrality = nx.closeness_centrality(H_cc)
print(f"Centralit√© de proximit√© calcul√©e pour {len(closeness_centrality)} n≈ìuds.")

top_close = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
for node, score in top_close:
    print(f"Node {node}: closeness = {score:.4f}")
```

## 4. PageRank

Le PageRank mesure l‚Äô**influence** d‚Äôun n≈ìud : un n≈ìud est important s‚Äôil est connect√© √† d‚Äôautres n≈ìuds importants.  
C‚Äôest la mesure la plus utilis√©e dans les graphes dirig√©s (mais reste informative ici aussi).


```python
pagerank = nx.pagerank(H_cc, alpha=0.85)
print(f"PageRank calcul√© pour {len(pagerank)} n≈ìuds.")

top_pr = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:5]
for node, score in top_pr:
    print(f"Node {node}: PageRank = {score:.4f}")
```

## 5. Centralit√© de Vecteur Propre (Eigenvector Centrality)

Cette mesure consid√®re qu‚Äôun n≈ìud est important s‚Äôil est connect√© √† d‚Äôautres n≈ìuds importants.  
C‚Äôest une version plus math√©matique du PageRank, bas√©e sur la d√©composition spectrale de la matrice d‚Äôadjacence.


```python
try:
    eigen_centrality = nx.eigenvector_centrality_numpy(H_cc)
except nx.NetworkXError:
    eigen_centrality = nx.eigenvector_centrality(H_cc, max_iter=500)

print(f"Centralit√© de vecteur propre calcul√©e pour {len(eigen_centrality)} n≈ìuds.")

top_eig = sorted(eigen_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
for node, score in top_eig:
    print(f"Node {node}: eigenvector = {score:.4f}")
```

## 4. Corr√©lation entre les centralit√©s et les attributs des utilisateurs

Maintenant que nous avons calcul√© les mesures de centralit√© pour chaque n≈ìud du graphe, nous allons :
1. Fusionner ces centralit√©s avec les attributs utilisateurs (√¢ge, genre, r√©gion, etc.).
2. Explorer leurs corr√©lations √† l‚Äôaide de statistiques simples (Pearson).
3. Identifier quels facteurs d√©mographiques ou comportementaux semblent li√©s √† la position d‚Äôun utilisateur dans le r√©seau.


```python
import pandas as pd

# Convertir les centralit√©s en DataFrames et fusionner
centralities_df = pd.DataFrame({
    "node": list(H_cc.nodes()),
    "degree": [degree_centrality[n] for n in H_cc.nodes()],
    "betweenness": [betweenness_centrality[n] for n in H_cc.nodes()],
    "closeness": [closeness_centrality[n] for n in H_cc.nodes()],
    "pagerank": [pagerank[n] for n in H_cc.nodes()],
    "eigenvector": [eigen_centrality[n] for n in H_cc.nodes()],
})

# Extraire les attributs d√©j√† attach√©s aux n≈ìuds
attrs = pd.DataFrame.from_dict(dict(H_cc.nodes(data=True)), orient="index").reset_index()
attrs.rename(columns={"index": "node"}, inplace=True)

# Fusion
merged = pd.merge(centralities_df, attrs, on="node", how="inner")
print(f"Merged dataset: {merged.shape[0]} nodes, {merged.shape[1]} columns")
merged.head()
```

### Corr√©lations num√©riques : √¢ge et completion
Nous testons ici la corr√©lation de Pearson entre les variables num√©riques (√¢ge, completion)
et chacune des mesures de centralit√©.


```python
from scipy.stats import pearsonr

num_cols = ["age", "completion"]
centrality_cols = ["degree", "betweenness", "closeness", "pagerank", "eigenvector"]

for y in centrality_cols:
    print(f"\n=== Corr√©lations avec {y} ===")
    for x in num_cols:
        valid = merged[[x, y]].dropna()
        r, p = pearsonr(valid[x], valid[y])
        print(f"{x:15s} r={r:.3f}  p={p:.4g}")
```

### Analyse cat√©gorielle : diff√©rence de centralit√© selon les attributs (ex : genre, r√©gion)
On compare maintenant les moyennes de centralit√© entre groupes √† l‚Äôaide de tests statistiques simples.


```python
cat_cols = ["gender", "region", "smoker_clean", "drinker_clean"]

for cat in cat_cols:
    if cat in merged.columns:
        print(f"\n===== {cat.upper()} =====")
        print(merged.groupby(cat)[centrality_cols].mean().round(3).sort_values("pagerank", ascending=False))
```

# √âtude comparative : mod√®les GNN pour pr√©dire la centralit√©

Cette section √©value plusieurs architectures de Graph Neural Networks (GCN, GraphSAGE, GAT)
pour **pr√©dire la centralit√© des n≈ìuds** √† partir des attributs d√©mographiques et de la structure du graphe Pokec.

Nous adoptons une m√©thodologie exp√©rimentale rigoureuse :
- **S√©paration des donn√©es** en ensembles train/validation/test.
- **Multiples ex√©cutions** avec seeds diff√©rentes pour estimer la variance.
- **M√©triques** : RMSE (erreur), R¬≤ (variance expliqu√©e), AUC (capacit√© √† d√©tecter les n≈ìuds les plus centraux).
- **√âtude d‚Äôablation** pour estimer l‚Äôimportance des attributs.


```python
# ============================================
# Conversion du graphe NetworkX vers PyTorch Geometric
# ============================================
import torch
from torch_geometric.utils import from_networkx

# Utiliser la composante connexe principale
G_nx = H_cc.copy()

# S'assurer que les indices de X correspondent bien aux noeuds du graphe
X_aligned = X.loc[list(G_nx.nodes())].sort_index()

# Conversion en tenseur
X_tensor = torch.tensor(X_aligned.to_numpy(), dtype=torch.float32)

# S√©lection d'une variable cible de centralit√© (par ex. closeness)
# V√©rifie d'abord que 'merged' contient bien les colonnes n√©cessaires
merged_aligned = merged.set_index("node").loc[list(G_nx.nodes())].sort_index()
y_tensor = torch.tensor(
    merged_aligned["closeness"].to_numpy(), dtype=torch.float32
).unsqueeze(1)

# Conversion NetworkX -> PyG
data_pyg = from_networkx(G_nx)
data_pyg.x = X_tensor
data_pyg.y = y_tensor

print(data_pyg)
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, SAGEConv, GATConv
from torch.nn import BatchNorm1d, Dropout


class GCN(torch.nn.Module):
    def __init__(self, in_dim, hidden, out_dim, dropout=0.3):
        super().__init__()
        self.conv1 = GCNConv(in_dim, hidden)
        self.norm1 = BatchNorm1d(hidden)
        self.conv2 = GCNConv(hidden, out_dim)
        self.drop = Dropout(dropout)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        x = self.norm1(x)
        x = self.drop(x)
        x = self.conv2(x, edge_index)
        return x

    def get_embeddings(self, x, edge_index):
        """Return hidden representations before the last layer."""
        x = F.relu(self.conv1(x, edge_index))
        x = self.norm1(x)
        return x.detach()


class GraphSAGE(torch.nn.Module):
    def __init__(self, in_dim, hidden, out_dim, dropout=0.3):
        super().__init__()
        self.conv1 = SAGEConv(in_dim, hidden)
        self.norm1 = BatchNorm1d(hidden)
        self.conv2 = SAGEConv(hidden, out_dim)
        self.drop = Dropout(dropout)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        x = self.norm1(x)
        x = self.drop(x)
        x = self.conv2(x, edge_index)
        return x

    def get_embeddings(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        x = self.norm1(x)
        return x.detach()


class GAT(torch.nn.Module):
    def __init__(self, in_dim, hidden, out_dim, heads=2, dropout=0.3):
        super().__init__()
        self.conv1 = GATConv(in_dim, hidden, heads=heads, dropout=dropout)
        self.norm1 = BatchNorm1d(hidden * heads)
        self.conv2 = GATConv(hidden * heads, out_dim, heads=1, dropout=dropout)
        self.drop = Dropout(dropout)

    def forward(self, x, edge_index):
        x = F.elu(self.conv1(x, edge_index))
        x = self.norm1(x)
        x = self.drop(x)
        x = self.conv2(x, edge_index)
        return x

    def get_embeddings(self, x, edge_index):
        x = F.elu(self.conv1(x, edge_index))
        x = self.norm1(x)
        return x.detach()
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error, roc_auc_score


# ---------------------------
# 1Ô∏è‚É£ TRAINING FUNCTION
# ---------------------------
def train_model(model, optimizer, data, epochs=150):
    """
    Train a GNN regression model on node-level centrality prediction.
    Handles shape mismatches and ensures sklearn compatibility.
    """
    model.train()
    for epoch in range(epochs):
        optimizer.zero_grad()
        out = model(data.x, data.edge_index).view(-1, 1)
        y = data.y.view(-1, 1)

        loss = F.mse_loss(out, y)
        loss.backward()
        optimizer.step()

        if epoch % 10 == 0:
            print(f"Epoch {epoch:03d} | Loss: {loss.item():.4f}")

    # Evaluation phase
    model.eval()
    with torch.no_grad():
        preds = model(data.x, data.edge_index).view(-1).cpu().numpy()
        y_true = data.y.view(-1).cpu().numpy()

        r2 = r2_score(y_true, preds)
        rmse = np.sqrt(mean_squared_error(y_true, preds))
        print(f"Final R¬≤: {r2:.3f} | RMSE: {rmse:.4f}")

    return model


# ---------------------------
# 2Ô∏è‚É£ WRAPPER TO TRAIN/EVAL MULTIPLE MODELS
# ---------------------------
def train_and_evaluate(model_class, name, lr=0.01, hidden=64, epochs=150):
    model = model_class(in_dim=data_pyg.num_features, hidden=hidden, out_dim=1)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=5e-4)
    train_model(model, optimizer, data_pyg, epochs=epochs)

    # Evaluate
    model.eval()
    with torch.no_grad():
        preds = model(data_pyg.x, data_pyg.edge_index).view(-1).cpu().numpy()
        y_true = data_pyg.y.view(-1).cpu().numpy()
        r2 = r2_score(y_true, preds)
        rmse = np.sqrt(mean_squared_error(y_true, preds))

        # AUC optional if you have binary task; fallback otherwise
        try:
            auc = roc_auc_score((y_true > np.median(y_true)).astype(int), preds)
        except ValueError:
            auc = np.nan

    results = pd.DataFrame([{"Model": name, "R¬≤": r2, "RMSE": rmse, "AUC": auc}])
    return results

```

```python
# Train and evaluate all three models
results_gcn = train_and_evaluate(GCN, "GCN")
results_sage = train_and_evaluate(GraphSAGE, "GraphSAGE")
results_gat = train_and_evaluate(GAT, "GAT")

# Comparison table
compare = pd.DataFrame({
    "Mod√®le": ["GCN", "GraphSAGE", "GAT"],
    "RMSE_moyen": [results_gcn["RMSE"].mean(), results_sage["RMSE"].mean(), results_gat["RMSE"].mean()],
    "R¬≤_moyen": [results_gcn["R¬≤"].mean(), results_sage["R¬≤"].mean(), results_gat["R¬≤"].mean()],
    "AUC_moyen": [results_gcn["AUC"].mean(), results_sage["AUC"].mean(), results_gat["AUC"].mean()],
}).round(4)

# Identify best-performing model
best_name = compare.loc[compare["R¬≤_moyen"].idxmax(), "Mod√®le"]
print(f"Best-performing model: {best_name}")

# Instantiate the best model using your argument names
if best_name == "GCN":
    best_model = GCN(in_dim=data_pyg.num_features, hidden=64, out_dim=1)
elif best_name == "GraphSAGE":
    best_model = GraphSAGE(in_dim=data_pyg.num_features, hidden=64, out_dim=1)
else:
    best_model = GAT(in_dim=data_pyg.num_features, hidden=64, out_dim=1)

# Retrain the chosen model on the full dataset
optimizer = torch.optim.Adam(best_model.parameters(), lr=0.01)
train_model(best_model, optimizer, data_pyg)

# Keep the original node features for embedding visualization
X_full = data_pyg.x.clone()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Similarit√© cosinus entre embeddings (√©chantillon pour √©viter les OOM)
sample_idx = np.random.choice(len(embeddings), size=500, replace=False)
emb_sample = embeddings[sample_idx]
meta_sample = merged.iloc[sample_idx][["region", "age_z"]]

sim_matrix = cosine_similarity(emb_sample)

# Visualisation : heatmap tri√©e par r√©gion
sorted_idx = np.argsort(meta_sample["region"])
sns.heatmap(sim_matrix[sorted_idx][:, sorted_idx], cmap="viridis", cbar=True)
plt.title("Similarit√© cosinus entre embeddings (tri√©e par r√©gion)")
plt.show()

# Moyenne de similarit√© intra/inter-r√©gions
meta_sample["region"] = meta_sample["region"].astype(str)
group_sim = []
for region in meta_sample["region"].unique():
    idx = meta_sample["region"] == region
    intra = sim_matrix[np.ix_(idx, idx)].mean()
    inter = sim_matrix[np.ix_(idx, ~idx)].mean()
    group_sim.append({"region": region, "intra": intra, "inter": inter})

pd.DataFrame(group_sim).sort_values("intra", ascending=False)
```

## Pr√©diction de liens ‚Äî test d‚Äôhomophilie

Nous ajoutons ici une t√¢che de **Link Prediction** :  
le mod√®le doit pr√©dire si deux utilisateurs devraient √™tre connect√©s.  
Cette exp√©rience permet de v√©rifier si la structure et les attributs des n≈ìuds
encodent bien des similarit√©s sociales (homophilie : "qui se ressemble se connecte").


```python
from torch_geometric.utils import negative_sampling
from sklearn.metrics import roc_auc_score

# Utiliser le bon mod√®le et les features complets
best_model.eval()
data_pyg.x = X_full  # on remet les features standardis√©s complets

with torch.no_grad():
    z = best_model(data_pyg.x, data_pyg.edge_index)

# Positives = ar√™tes existantes
pos_edge_index = data_pyg.edge_index

# N√©gatives = ar√™tes absentes (√©chantillonn√©es)
neg_edge_index = negative_sampling(
    pos_edge_index,
    num_nodes=data_pyg.num_nodes,
    num_neg_samples=pos_edge_index.size(1),
)

# Fonction utilitaire : produit scalaire entre embeddings de deux n≈ìuds
def link_logits(z, edge_index):
    return (z[edge_index[0]] * z[edge_index[1]]).sum(dim=-1)

# Calcul des scores pour les liens existants / absents
with torch.no_grad():
    pos_score = torch.sigmoid(link_logits(z, pos_edge_index)).cpu().numpy()
    neg_score = torch.sigmoid(link_logits(z, neg_edge_index)).cpu().numpy()

# Labels + Pr√©dictions concat√©n√©es
y_true = np.concatenate([
    np.ones_like(pos_score),
    np.zeros_like(neg_score)
])
y_pred = np.concatenate([pos_score, neg_score])

auc = roc_auc_score(y_true, y_pred)
print(f"AUC (homophilie des liens) : {auc:.4f}")
```
